# Feature Engineering

```python
#As we don't have null values now we have to handle the cateogrical_features
```

### Handling Categorical features

```python
categorical_features
```

```python
df.drop(['CarName'],axis=1,inplace=True)#car name is not necessary for predicting the price of the car
```

```python
df
```

```python
categorical_features.remove("CarName")
for feature in categorical_features:
    print(feature,len(df[feature].unique()),df[feature].unique())
```

```python
# As fueltype Nominal categorical feature, so we have to do nominal encoding
fueltype=df[['fueltype']]
fueltype=pd.get_dummies(fueltype,drop_first=True)
```

```python
fueltype
```

```python
# As aspiration is also Nominal categorical feature,so we have to do nominal encoding
aspiration=df[['aspiration']]
aspiration=pd.get_dummies(aspiration,drop_first=True)
```

```python
aspiration
```

```python
#As doornumber is label categorical feature we have to do label encoding 
df['doornumber'].replace({'two':2,'four':4},inplace=True)
```

```python
df
```

```python
# As carbody is also a nominal categorical feature we have to apply nominal encoding
carbody=df[['carbody']]
carbody=pd.get_dummies(carbody,drop_first=True)
```

```python
carbody
```

```python
#Drivewheel is also nominal categorical feature
drivewheel=df[['drivewheel']]
drivewheel=pd.get_dummies(drivewheel,drop_first=True)
```

```python
drivewheel
```

```python
## engine location is a nominal categorical feature
enginelocation=df[['enginelocation']]
enginelocation=pd.get_dummies(enginelocation,drop_first=True)
```

```python
enginelocation
```

```python
#enginetype-Nominal categorical feature
enginetype=df[['enginetype']]
enginetype=pd.get_dummies(enginetype,drop_first=True)
```

```python
enginetype
```

```python
#cylinder number is label categorical feature
df['cylindernumber'].replace({'two':2,'four':4,'six':6,'three':3,'five':5,'twelve':12,'eight':8},inplace=True)
```

```python
df['cylindernumber']
```

```python
#Fuel system is nominal categorical feature
fuelsystem=df[['fuelsystem']]
fuelsystem=pd.get_dummies(fuelsystem,drop_first=True)
```

```python
fuelsystem
```

```python
df=pd.concat([df,fueltype,aspiration,carbody,drivewheel,enginelocation,enginetype,fuelsystem],axis=1)
```

```python
df
```

```python
df.drop(['fueltype','aspiration','carbody','drivewheel','enginelocation','enginetype','fuelsystem'],axis=1,inplace=True)
```

```python
df
```

```python
df.columns
```

```python
X=df.drop(labels=['price'],axis=1)
y=df['price']
```

```python
X
```

```python
y
```

# Training the model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.3)
```

```python
print(X_train.shape,X_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
rf.score(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
from sklearn import metrics
import numpy as np
print('MAE : ',metrics.mean_absolute_error(y_test,y_pred))
print('MSE : ',metrics.mean_squared_error(y_test,y_pred))
print('RMSE : ',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
metrics.r2_score(y_test,y_pred)
```
